<!-- Describe how this figure helps address your research question. What patterns or insights can you observe? -->
**Figure 1 Caption:** The vast majority of dots on this map have a darker color, indicating that most eruptions in the last 1,000 years had a lower VEI, which helps to support my hypothesis that explosive eruptions are not as common as effusive (non explosive) eruptions.
